```python
import pandas as pd
import numpy as np

data = pd.read_excel("/Users/bahus/Documents/programing/Diploma work2/comments.xlsx")
```

```python
data = data.astype(str)
```

```python
df = data[~data['Comments'].str.startswith('@')]
```

```python
is_emoji = df.apply(lambda row: row.str.match(r'^[\U0001F300-\U0001F5FF\U0001F600-\U0001F64F\U0001F680-\U0001F6FF\U0001F700-\U0001F77F\U0001F780-\U0001F7FF\U0001F800-\U0001F8FF\U0001F900-\U0001F9FF\U0001FA00-\U0001FA6F\U0001FA70-\U0001FAFF\U00002702-\U000027B0\U000024C2-\U0001F251]+$')).all(axis=1)

# Drop rows containing only emojis
df = df[~is_emoji]
```

```python
from langdetect import detect

# Identify languages and drop Russian and Kazakh comments
filtered_comments = [comment for comment in df['Comments'] if detect(comment) not in ['ru', 'kk']]

# Print filtered comments
print(filtered_comments)
```
