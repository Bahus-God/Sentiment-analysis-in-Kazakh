```python
# Import necessary libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from transformers import BertTokenizer, BertModel
import torch
import numpy as np

# Step 2: Preprocessing
def preprocess_data(data):
    data = pd.DataFrame(data['Comments'])
    return data

# Step 3: Feature Extraction
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-uncased')
model = BertModel.from_pretrained('bert-base-multilingual-uncased')

def extract_features(texts):
    features = []
    for text in texts:
        inputs = tokenizer(text, return_tensors='pt')
        outputs = model(**inputs)
        features.append(outputs.last_hidden_state[:,0,:].detach().numpy())
    return np.vstack(features)

# Step 4: Clustering
def cluster_texts(features, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init = 10).fit(features)
    return kmeans.labels_

# Load and preprocess data
preprocessed_data = preprocess_data(data)

# Extract features and cluster texts
features = extract_features(preprocessed_data)
labels = cluster_texts(features, n_clusters=1)

# Now each text in your data is assigned a cluster label from 0 to 4
```
